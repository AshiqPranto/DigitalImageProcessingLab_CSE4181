```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/tiger.jpeg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

Convertion from Gray Image to Binary Image


```python
threshold_value = 128
binary_img = np.where(img_gray > threshold_value, 1, 0)

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.title("Original Gray Image")
plt.imshow(img_gray, cmap="gray")

plt.subplot(2, 2, 2)
plt.title("Converted Binary Image")
plt.imshow(binary_img, cmap="gray")


```

Convertion from Gray Image to Negative Image


```python
max_intensity = 255
negative_img = max_intensity - img_gray

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.title("Original Gray Image")
plt.imshow(img_gray, cmap="gray")

plt.subplot(2, 2, 2)
plt.title("Converted Negative Image")
plt.imshow(negative_img, cmap="gray")
```
