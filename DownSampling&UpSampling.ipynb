```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/tiger.jpeg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
# print(img.shape)
plt.figure(figsize=(10, 16))

plt.subplot(2, 2, 1)
plt.title("RGB")
plt.imshow(img_rgb)
plt.subplot(2, 2, 2)
plt.title("BGR")
plt.imshow(img_bgr)
plt.subplot(2, 2, 3)
plt.title("Gray")
plt.imshow(img_gray, cmap="gray")
```

```python
height, width = img_gray.shape
# Downsample the image by a factor of 30 with using built in function 
downsampled_img = cv2.resize(img_gray, (img_gray.shape[1] // 30, img_gray.shape[0] // 30))

# Downsample the image by a factor of 30 without using built in function
new_width = width // 30
new_height = height // 30
downsampled_img1 = np.zeros((new_height, new_width), dtype=np.uint8)

# Perform downsampling by averaging (new_width*new_height) elements
for i in range(new_height):
    for j in range(new_width):
        downsampled_img1[i][j] = np.mean(img_gray[i*30 : ((i*30)+30), j*30 : ((j*30)+30)])

plt.figure(figsize=(10, 16))

print(downsampled_img.shape)
# Display the downsampled image
plt.subplot(1, 2, 1)
plt.title("DownSampling Using built in function")
plt.imshow(downsampled_img,cmap="gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("DownSampling Without built in function")
plt.imshow(downsampled_img1, cmap='gray')
plt.axis("off")

```

```python
# Upsampling the image by a factor of 30 with using built in function
upsampled_img = cv2.resize(downsampled_img, (downsampled_img.shape[1] * 30, downsampled_img.shape[0] * 30), interpolation=cv2.INTER_NEAREST)

plt.figure(figsize=(10, 16))
print(upsampled_img.shape)
plt.subplot(1, 2, 1)
plt.title("UpSampled using built in function")
plt.imshow(upsampled_img,cmap="gray")
plt.axis("off")

# Upsampling the image without using any builtin function
new_height = downsampled_img.shape[0] * 30
new_width = downsampled_img.shape[1] * 30
upsampled_img1 = np.zeros((new_height, new_width), dtype=np.uint8)

for i in range(new_height):
    for j in range(new_width):
        upsampled_img1[i][j] = downsampled_img1[(i // 30)][(j // 30)]

plt.subplot(1, 2, 2)
plt.title("UpSampled without built in function")
plt.imshow(upsampled_img1, cmap='gray')
plt.axis("off")

```
