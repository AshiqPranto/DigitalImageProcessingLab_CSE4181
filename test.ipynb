```python
height, width = img.shape[:2]
midPoint_height = height//2
midPoint_width = width//2
top_left = img[0:midPoint_height, 0:midPoint_width]
top_right = img[0: midPoint_height, midPoint_width:width]
bottom_left = img[midPoint_height:height, 0:midPoint_width]
bottom_right = img[midPoint_height:height, midPoint_width:width]
plt.subplot(2,2,1)
plt.imshow(top_left)
plt.subplot(2,2,2)
plt.imshow(top_right)
plt.subplot(2,2,3)
plt.imshow(bottom_left)
plt.subplot(2,2,4)
plt.imshow(bottom_right)
```

```python

reconstructed_img = np.vstack((np.hstack((top_left, top_right)), np.hstack((bottom_left, bottom_right))))
plt.imshow(reconstructed_img)
```
