```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/tiger.jpeg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_gray = cv2.resize(img_gray, (512, 512))
```

```python
# Display the original image
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Image")
plt.axis("off")
# plt.show()

lower_limit = 0
upper_limit = 50

# Perform brightness enhancement within the specified range
enhanced_img = img_gray
enhanced_img[(enhanced_img >= lower_limit) & (enhanced_img <= upper_limit)] += 200

# Display the enhanced image
plt.subplot(1, 2, 2)
plt.imshow(enhanced_img, cmap='gray')
plt.title("Enhanced Image")
plt.axis("off")
plt.show()
```
