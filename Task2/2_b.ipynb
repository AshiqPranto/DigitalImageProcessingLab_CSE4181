```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "../SrcImage/sample1.jpg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img = cv2.resize(img_gray, (512, 512))
```

```python
import cv2
import numpy    as np
import matplotlib.pyplot as plt

# Define gamma and c for the power law transform
gamma = 0.4
c = 100
plt.hist(img.ravel(), 256)
# Apply the power law transform
power_transformed = c * np.power(img, gamma)

# Apply the logarithmic transform
# log_transformed = c * np.log(1 + img)
log_transformed = c * (255 * (np.log(1 + img/255) / np.log(2)))
plt.figure(figsize=(8, 8))
plt.hist(log_transformed.ravel(), 256)
# Apply the inverse logarithmic transform
c = 255
inverse_log_transformed = 255 * (np.power(2, (img/c)) - 1)


# Display the original, power transformed, and inverse logarithmic transformed images
plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# plt.figure(figsize=(12, 4))
plt.subplot(2, 2, 2)
plt.imshow(log_transformed, cmap='gray')
plt.title("log transformed Image")
plt.axis("off")



plt.subplot(2, 2, 3)
plt.imshow(power_transformed, cmap='gray')
plt.title(f"Power Law Transform gamma = {gamma}")
plt.axis("off")

plt.subplot(2, 2, 4)
# plt.imshow(cv2.cvtColor(inverse_log_transformed, cv2.COLOR_BGR2RGB))
plt.imshow(inverse_log_transformed, cmap='gray')
plt.title("Inverse Logarithmic Transform")
plt.axis("off")

plt.show()
```
