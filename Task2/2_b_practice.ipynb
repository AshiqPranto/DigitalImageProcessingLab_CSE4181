```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "../SrcImage/lenna.png"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_gray = cv2.resize(img_gray, (512, 512))
```

```python
import numpy as np

gamma = 0.4
c = 100
power_transformed = c * (img_gray ** gamma)

log_transformed = c * np.log(1 + img_gray)

inverse_log_transformed = 255 * ((2 ** (img_gray / c)) - 1)

plt.subplot(2, 2, 1)
plt.title("Original image")
plt.imshow(img_gray, cmap='gray')

plt.subplot(2, 2, 2)
plt.title("Power transform image")
plt.imshow(power_transformed, cmap='gray')

plt.subplot(2, 2, 3)
plt.title("Log transformed")
plt.imshow(log_transformed, cmap='gray')

plt.subplot(2, 2, 4)
plt.title("Anti Log transformed")
plt.imshow(inverse_log_transformed, cmap='gray')
```
