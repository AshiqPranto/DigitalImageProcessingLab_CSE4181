```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/tiger.jpeg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_gray = cv2.resize(img_gray, (512, 512))
```

```python
msb_img = img_gray >> 5
difference_img = img_gray - msb_img

plt.figure(figsize=(8, 10))
#Plot the original image
plt.subplot(1, 3, 1)
plt.axis("off")
plt.imshow(img_gray, cmap="gray")
plt.title("Original Image")

# plot the MSB image
plt.subplot(1, 3, 2)
plt.axis("off")
plt.imshow(msb_img, cmap="gray")
plt.title("MSB Image")

# plot the difference image
plt.subplot(1, 3, 3)
plt.axis("off")
plt.imshow(difference_img, cmap="gray")
plt.title("Difference Image")
```
