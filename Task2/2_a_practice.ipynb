```python
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = "../SrcImage/lenna.png"
img_bgr = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_gray = cv2.resize(img_gray, (512, 512))
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original image")
plt.axis("off")

lower_limit = 0
upper_limit = 50

#Perform brightness enhancement within the specified range
enhanced_img = img_gray
enhanced_img[(enhanced_img >= lower_limit) & (enhanced_img <= upper_limit)] += 200


plt.subplot(1, 2, 2)
plt.imshow(enhanced_img, cmap='gray')
plt.title("Enhanced Image")
```
