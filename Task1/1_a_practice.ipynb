```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path = "../SrcImage/flower.jpg"
img_path = "../SrcImage/lenna.png"
img_bgr = cv2.imread(img_path)
img_bgr = cv2.resize(img_bgr, (512, 512))
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(img_bgr)
plt.subplot(1, 3, 2)
plt.imshow(img_rgb)
plt.subplot(1, 3, 3)
plt.imshow(img_gray, cmap="gray")
```

```python
def decrease_resolution_by_2(sample_img):
    height, width = sample_img.shape
    new_height = height // 2
    new_width = width // 2
    decreased_img = np.zeros((new_height, new_width))

    for i in range(new_height):
        for j in range(new_width):
            decreased_img[i][j] = np.mean(sample_img[i*2 : (i*2)+2, (j*2) : (j*2 + 2)])

    return decreased_img
```

```python
reduced_img = img_gray.copy()
plt.figure(figsize=(16, 10))
for cnt in range(1,6):
    plt.subplot(2, 3, cnt)
    plt.title(f"{reduced_img.shape}")
    plt.imshow(reduced_img)
    reduced_img = decrease_resolution_by_2(reduced_img)
```
