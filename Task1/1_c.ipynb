```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/sketch5.jpg"  # Replace with your image path
img_rgb = plt.imread(path)
img_gray = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
img_gray = cv2.resize(img_gray,(512,512))

plt.figure(figsize=(8, 6))
plt.hist(img_gray.ravel(), bins=256)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()

# Determine the threshold value from the Histogram
threshold_value = 58

# Apply thresholding to perform segmentation
segmented_img = np.where(img_gray >= threshold_value, 255, 0)

# Display the original and segmented images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(segmented_img, cmap='gray')
plt.title("Segmented Image")
plt.axis("off")

plt.show()
```
