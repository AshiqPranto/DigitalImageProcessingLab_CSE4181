```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Read the original image
path = "/media/mount/nvme0n1p4/Pranto/study/4-1/Digital Image Processing/DIP_Lab/SrcImage/tiger.jpeg"  # Replace with your image path
img_bgr = cv2.imread(path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
# print(img.shape)
print(img_gray.shape)
img_gray = cv2.resize(img_gray, (512, 512))
print(img_gray.shape)
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title("RGB")
plt.imshow(img_rgb)
plt.subplot(2, 2, 2)
plt.title("BGR")
plt.imshow(img_bgr)
plt.subplot(2, 2, 3)
plt.title("Gray")
plt.imshow(img_gray, cmap="gray")
```

```python
import time

def decrease_resolution_by_2(sample_img):
    height, width = sample_img.shape
    new_height = height // 2
    new_width = width // 2
    downsampled_img1 = np.zeros((new_height, new_width))
    # downsampled_img1 = cv2.resize(sample_img,(new_width, new_height))
    # Perform downsampling by averaging (new_width*new_height) elements
    for i in range(new_height):
        for j in range(new_width):
            downsampled_img1[i][j] = np.mean(sample_img[i*2 : ((i*2)+2), j*2 : ((j*2)+2)])
    return downsampled_img1
cnt = 1
while cnt < 5:
    print(cnt)
    img_gray = decrease_resolution_by_2(img_gray)
    title = "Down Sampling " + str(cnt) + " times"
    plt.title(title)
    plt.imshow(img_gray, cmap='gray')
    plt.show()
    cnt = cnt + 1
    # time.sleep(1)
```
