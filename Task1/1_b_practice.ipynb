```python
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = "../SrcImage/Skull_374x452.tif"

img_bgr = cv2.imread(img_path)
img_bgr = cv2.resize(img_bgr, (512, 512))
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(16, 8))
# plt.imshow(img_gray)

total_bits = 8
for current_bit in range(0, total_bits):
    div = 2 ** current_bit
    reduced_img = (img_gray // div) * div
    plt.subplot(2, 4, current_bit+1)
    plt.imshow(reduced_img, cmap="gray")
```
